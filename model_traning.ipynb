```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
```

```python
def calculate_ema(prices, days, smoothing=2):
    ema = [sum(prices[:days]) / days]
    for price in prices[days:]:
        ema.append((price * (smoothing / (1 + days))) + ema[-1] * (1 - (smoothing / (1 + days))))
    return ema
```

```python
""" ##İnputs##
smaFastPer= 3
smaMidPer =10
smaSLowPer =15
symbol ='BTC-USD' """

##İnputs##
smaFastPer= 3
smaMidPer =10
smaSLowPer =14
STDsp = 3
symbol ='BTC-USD'


```

```python

df = web.DataReader(symbol, 'yahoo', '2019-01-01', '2022-01-01')

##statical calculates

#ema = calculate_ema(df['Close'], 10)
df['pandas_SMA_fast'] = df.iloc[:,3].rolling(window=smaFastPer).mean()
df['pandas_SMA_mid'] = df.iloc[:,3].rolling(window=smaMidPer).mean()
df['pandas_SMA_slow'] = df.iloc[:,3].rolling(window=smaSLowPer).mean()

#bollinger
rstd = df.iloc[:,3].rolling(window=smaSLowPer).std()

upper_band = df['pandas_SMA_slow'] + STDsp * rstd

lower_band = df['pandas_SMA_slow'] - STDsp * rstd

df['upper_band'] = upper_band
df['lower_band'] = lower_band

#RSI####

delta = df['Close'].diff()
up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)
ema_up = up.ewm(com=smaSLowPer, adjust=False).mean()
ema_down = down.ewm(com=smaSLowPer, adjust=False).mean()
rs = ema_up/ema_down

df['RSI'] = 100 - (100/(1 + rs))

df
```

```python
df_filled = df.dropna(subset=['pandas_SMA_slow'])
df_filled.isnull().sum()
```

```python
#lag price##
dflag=df_filled
""" dflag['LAGpandas_SMA_slow'] =dflag['pandas_SMA_slow'].shift(-1)
dflag['LAGpandas_SMA_fast'] =dflag['pandas_SMA_fast'].shift(-1)
dflag['LAGpandas_SMA_mid'] =dflag['pandas_SMA_mid'].shift(-1) """
#tagret fiyatı belirlemek için bir önceki high'ı yanına getirdik
dflag['lag1High'] =dflag['High'].shift(1)
dflag
```

```python
"""dflagdropna = dflag.dropna(subset=['LAGpandas_SMA_slow'])"""
dflagdropna = dflag.dropna(subset=['lag1High'])
dflagdropna.isnull().sum()
```

```python
dflagdropna
```

```python
dflagdropna['high_close'] = dflagdropna['lag1High']- dflagdropna['Close']
#dflagdropna['highround'] = round(dflagdropna['High'], 2)
dflagdropna= round(dflagdropna, 2)
dflagdropna
```

```python
#değişşim hesaplama
dflagdropna['Degisim'] = (dflagdropna['high_close']/dflagdropna['Close'])*100
dflagdropna
```

```python
#conditions or strategy
dflagdropna.loc[dflagdropna['Degisim'] <= 1.50, 'TagretProfit'] = '0' 
dflagdropna.loc[dflagdropna['Degisim'] > 1.50, 'TagretProfit'] = '1' 
"""is it green bar?"""
dflagdropna.loc[dflagdropna['Close'] > dflagdropna['Open'] , 'GreenBar'] = '1' 
dflagdropna.loc[dflagdropna['Close'] <= dflagdropna['Open'] , 'GreenBar'] = '0' 

"""is Close Price over the SMA? """
#fast
dflagdropna.loc[dflagdropna['Close'] > dflagdropna['pandas_SMA_fast'] , 'overFastSMA'] = '1' 
dflagdropna.loc[dflagdropna['Close'] <= dflagdropna['pandas_SMA_fast'] , 'overFastSMA'] = '0' 
#mid
dflagdropna.loc[dflagdropna['Close'] > dflagdropna['pandas_SMA_mid'] , 'overMidSMA'] = '1' 
dflagdropna.loc[dflagdropna['Close'] <= dflagdropna['pandas_SMA_mid'] , 'overMidSMA'] = '0' 
#slow
dflagdropna.loc[dflagdropna['Close'] > dflagdropna['pandas_SMA_slow'] , 'overSlowSMA'] = '1' 
dflagdropna.loc[dflagdropna['Close'] <= dflagdropna['pandas_SMA_slow'] , 'overSlowSMA'] = '0' 

#bollinger
#upper
dflagdropna.loc[dflagdropna['Close'] > dflagdropna['upper_band'] , 'bolUpper'] = '1' 
dflagdropna.loc[dflagdropna['Close'] <= dflagdropna['upper_band'] , 'bolUpper'] = '0' 
#lower
dflagdropna.loc[dflagdropna['Close'] > dflagdropna['lower_band'] , 'bolLower'] = '1' 
dflagdropna.loc[dflagdropna['Close'] <= dflagdropna['lower_band'] , 'bolLower'] = '0' 

#RSI
#70updown
dflagdropna.loc[dflagdropna['RSI'] > 70 , 'RSI70'] = '1' 
dflagdropna.loc[dflagdropna['RSI'] <= 70 , 'RSI70'] = '0' 
#30updown
dflagdropna.loc[dflagdropna['RSI'] > 30 , 'RSI30'] = '0' 
dflagdropna.loc[dflagdropna['RSI'] <= 30 , 'RSI30'] = '1' 


dflagdropna
```

```python
## Creating Dataframe For Traning Model##
TrainDataFrame = dflagdropna[['GreenBar','overMidSMA','overFastSMA','overMidSMA','overSlowSMA','bolUpper','bolLower','RSI70','RSI30','TagretProfit']]
TrainDataFrame
```

```python
#libraries
from numpy import loadtxt
from xgboost import XGBClassifier,XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import model_selection, tree, preprocessing, metrics, linear_model

```

```python
#train test split
X =  TrainDataFrame.drop(['TagretProfit'], axis = 1)
y = TrainDataFrame['TagretProfit']
seed = 7
test_size = 0.33

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=test_size, random_state=seed)
X_train
```

```python
X_train.dtypes
```

```python
def fit_ml_algo(algo, X_train, y_train, cv):
    
    # One Pass
    model = algo.fit(X_train, y_train)
    acc = round(model.score(X_train, y_train) * 100, 2)
    
    # Cross Validation 
    train_pred = model_selection.cross_val_predict(algo, 
                                                  X_train, 
                                                  y_train, 
                                                  cv=cv, 
                                                  n_jobs = -1)
    # Cross-validation accuracy metric
    acc_cv = round(metrics.accuracy_score(y_train, train_pred) * 100, 2)
    
    return train_pred, acc, acc_cv
```

```python
# Decision Tree Classifier

train_pred_dt, acc_dt, acc_cv_dt = fit_ml_algo(DecisionTreeClassifier(), 
                                                                X_train, 
                                                                y_train,
                                                                10)

print("Accuracy: %s" % acc_dt)
print("Accuracy CV 10-Fold: %s" % acc_cv_dt)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
